```
import math as m
import numpy as np

Pi = 3.14
Sqrt2 = m.sqrt(2)
Funtion = lambda x: m.sin(x)

def AddMultiply(Matrix, ChangingRow, Row, Value):
  NewRow = Matrix[ChangingRow,:]+Value*Matrix[Row,:]
  Matrix2 = []
  for i in range(len(Matrix)):
    if i != ChangingRow:
      Matrix2.append(Matrix[i,:])
    else:
      Matrix2.append(NewRow)
  return np.array(Matrix2)

def LeadingOne(Matrix,Row,Column):
  Matrix2 =[]
  for i in range(len(Matrix)):
    if i != Row:
      Matrix2.append(Matrix[i,:])
    else:
      Matrix2.append(1/Matrix[Row,Column]*Matrix[Row,:])
  return np.array(Matrix2)

def GaussJordan(Matrix):                                                              #performing Gauss-Jordan process on leading one
  Matrix2 = Matrix
  for Pivot in range(len(Matrix)):
    Matrix2 = LeadingOne(Matrix2,Pivot,Pivot)
    for Column in range(0,Pivot):
      Matrix2 = AddMultiply(Matrix2,Column,Pivot,-Matrix2[Column,Pivot])
    for Column in range(Pivot+1,len(Matrix)):
      Matrix2 = AddMultiply(Matrix2,Column,Pivot,-Matrix2[Column,Pivot])
  return Matrix2
```

```
def NewRow(Input, Output, Order):
  Row = []
  for i in range(Order + 1):
    Row.append(Input**i)
  Row.append(Output)
  return Row
```

```
def NewMatrix(DataPoints):
  Order = len(DataPoints) - 1
  DataLen = len(DataPoints)
  Matrix = []
  for i in range(DataLen):
    z = NewRow(DataPoints[i][0], DataPoints[i][1], Order)
    Matrix.append(z)
  return np.array(Matrix)
```

Interpolation is a way to find the value of y at any other value of x, given (x0,y0), (x1,y1),………….....…...(xn,yn). Polynomials are the most commonly choice for interpolants because they are easy to use evaluate, differentiate and integrate. Polynomial interpolation involves finding a polynomial of order ‘n’ that passes through the ‘n+1’ points. Direct method is one of the ways to find the polynomial interpolation along with some other ways such as the Lagrangian interpolation method, Newton’s divide polynomial method. The direct method of interpolation is based on the following premise. Given, ‘n+1’ data points (x0,y0), (x1,y1),……………..….(xn,yn), pass a polynomial of order ‘n’ through the data as given below: 
y=a0 + a1 x+……………………+anxn [: - where a0, a1, ……… an are real constants]. 
Here, you can setup ‘n+1’ equations to find ‘n+1’ constants and to find the value of ‘x’, simply substitute the value of ‘x’ in the above polynomial.
