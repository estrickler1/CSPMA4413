```
def TrueError(Estimate, Actual):                                                      #Defines True Error
  return Actual - Estimate

def RelativeTrueError(Estimate, Actual):                                              #Defines Relative True Error
  return TrueError(Estimate, Actual) / Actual

def ApproximateError(PreEstimate, CurEstimate):                                       #Defines Approximate Error
  return CurEstimate - PreEstimate

def RelativeApproximateError(PreEstimate, CurEstimate):                               #Defines Relative Approximate Error
  return ApproximateError(PreEstimate, CurEstimate) / CurEstimate
```

```
import math

Pi = math.pi                                                                           #Number Pi
Sqrt2 = math.sqrt(2)                                                                   #Number SQRT(2)
Center = Pi/4                                                                          #Taylor Series Center

def GeneralTaylorCoefficient(Degree):                                                  #Defines General Taylor Series
  return (-1)**(math.floor(Degree/2))/(Sqrt2*math.factorial(Degree))

def TaylorPolynomialTerm(Degree, Input):                                               #Defines Specific Term in Taylor Series
  return GeneralTaylorCoefficient(Degree)*(Input-Center)**Degree

def TaylorPolynomial(Degree, Input):                                                   #Defines a Taylor Series, with a Specified Order, and Input
  sum = 0
  for i in range(0, Degree+1):
    sum = sum +  TaylorPolynomialTerm(i, Input)
  return sum
```

```
Degree = [i for i in range(0,31)]                                                      #Order of Taylor Series
Actual = [math.sin(Center+.1) for i in Degree]                                         #Actual Value
Estimate = [TaylorPolynomial(n,Center+.1) for n in Degree]                             #Taylor Series Estimation
TrueE = [TrueError(Estimate[i],Actual[i]) for i in Degree]                             #True Error
RelTruE = [RelativeTrueError(Estimate[i],Actual[i]) for i in Degree]                   #Relative True Error
ApxE = [ApproximateError(Estimate[i],Estimate[i+1]) for i in range(0,30)]              #Approximate Error
RelApxE = [RelativeApproximateError(Estimate[i],Estimate[i+1]) for i in range (0,30)]  #Relative Approximate Error
```

```
import plotly.graph_objects as go

fig = go.Figure(data=[go.Table(
                header=dict(values=['Degree', 'Actual Value', 'Estimated Value','True Error','Relative True Error','Approximate Error','Relative Approximate Error']),
                 cells=dict(values = [Degree,Actual,Estimate,TrueE,RelTruE,ApxE,RelApxE]))
                     ])
fig.show()                                                                                #Shows the Created Table
```

The purpose for this program is to explore different types of errors, by using an estimated value from a Taylor Series. The errors being looked into are true errors, relative true errors, approximate errors, and relative approximate errors. The following Taylor Series is used to create an estimation of sin(x),
$$
f_n(x) = \sum_{i=0}^n  \frac{\left(-1\right)^{\lfloor\frac n2\rfloor} }{\sqrt 2\cdot n!}
\left(x-\frac \pi 4\right)^n
$$

By finding an estimation value, the true error is calculated by subtracting the estimation from the actual value. Then the relative true error is found by dividing the true error by the actual value. 

$$
True Error = Actual Value - Estimated Value
$$


$$
Relative True Error = \frac{True Error}{Actual Value}
$$

To find the approximate error and the relative approximate error, the order of the Taylor Series is very important. Each of these error uses the previous estimate and the current estimation. To find the approximate error, previous estimation is subtracted from the current estimate. Then to find the relative approximate error, the approximate error is divided by the current estimation.

$$
Approximate Error = Current Estimation - Previous Estimation
$$


$$
Relative Approximate Error = \frac{Approximate Error}{Current Estimation}
$$
